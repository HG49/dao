```python
def ball_euler(t0, tend, y0, v0, h):    
    t, y, v, i = [t0], [y0], [v0], 0
    a = -9.8     
    while t[-1] <= tend and y[-1] > 0:
        y.append(y[-1] + h*v[-1])
        v.append(v[-1] + h*a)
        i += 1
        t.append(i*h)        
    return np.array(t), np.array(y), np.array(v)
```
